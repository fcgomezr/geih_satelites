```python
import os
import pandas as pd
import numpy as np
from rich import print
from MT_job import MT_job as mt
```

```python

year = 2023
#os.chdir('E:/GEIH/Empalme/'+str(year)+'/')
path = r'E:/GEIH/Empalme/Consolidada/'
df = pd.read_csv(path+'GEIH_'+str(year)+'_'+'.csv',sep=",",parse_dates=True,na_values=['nan'])
ciiu = pd.read_csv(path+'CIIU.csv',sep=";",parse_dates=True,na_values=['nan'])
df = pd.merge(df, ciiu, on=['RAMA4D_R4'], how='left')
print("Current working directory: {0}".format(os.getcwd()))
```

```python
df.groupby(['PER', 'MES'])[['FEX_C18','horas_efectivas']].sum().reset_index().groupby(['PER'])[['FEX_C18','horas_efectivas']].mean().reset_index()

db = pd.pivot_table(df,
               values=['FEX_C18','horas_efectivas','asal','ind','asal_hours','ind_hours'], 
               index=['RAMA4D_R4'],
               columns=['MES'], 
               aggfunc=np.sum, fill_value=0)
table2 = db.groupby(axis=1, level=0).mean()

#Read excel 
ciiu_s = pd.read_excel('CIIU_CSTIC.xlsx',header=0,index_col=0)

# Merge ciiu_s with table2 llaves RAMA4D_R4 con CIIU 4
merged_df = pd.merge(ciiu_s, table2, left_on='CIIU 4',right_on='RAMA4D_R4', right_index=True)
merged_df['PER'] =year
merged_df
#Guardar en excel
```
